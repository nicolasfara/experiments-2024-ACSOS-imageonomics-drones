```kotlin
fun linearizationFunction(value: Double, min: Double, max: Double) = when {
    value >= max -> 1.0 // If  value is greater than MAX then fix it to 1.0
    else -> (value - min) / (max - min) // Change range into [0..1]
}

fun sigmoid(value: Double, m: Double, v: Double): Double {
    return (1 + ((value * (1 - m)) / (m * (1 - value))).pow(-v)).pow(-1)
}

val values = (20..80).toList()
val n5 = values.map { sigmoid(
    linearizationFunction(it.toDouble(), 20.0, 80.0),
    0.35,
    5.0
) }

val n10 = values.map { sigmoid(
    linearizationFunction(it.toDouble(), 20.0, 80.0),
    0.35,
    10.0
) }

val nBetter = values.map { sigmoid(
    linearizationFunction(it.toDouble(), 20.0, 80.0),
    1.0 / 3.0,
    4.0
) }

plot {
    line {
        x(values)
        y(n5)
        color = Color.PURPLE
    }

    line {
        x(values)
        y(n10)
        color = Color.BLUE
    }

    line {
        x(values)
        y(nBetter)
        color = Color.YELLOW
        width = 3.0
    }
    
    line {
        y(0..1)
        x.constant(40.0)
        type = LineType.DOTTED
        width = 2.0
    }
    
    layout { // Set plot layout
        title = "Simple plot with kandy-lets-plot" // Add title
        // Add caption
        caption = "See `examples` section for more\n complicated and interesting examples!"
        size = 2000 to 1500 // Plot dimension settings
    }
}
```
